##Data Preprocessing

```python
del weather_data['Unnamed: 0']
```

```python
weather_data.head()
```

Client has agreed for mean imputation for ozone

```python
ozone_mean = weather_data['Ozone'].mean()
```

```python
weather_data['Ozone'].fillna(value= ozone_mean,inplace =True)
```

```python
weather_data.isnull().sum()
```

Client has agreed for median imputation for solar.R

```python
solar_median = weather_data['Solar.R'].median()
```

```python
weather_data['Solar.R'].fillna(value= solar_median,inplace =True)
```

```python
weather_data.isnull().sum()
```

```python
weather_data['Weather'].unique()
```

```python
#As we can see that there are 3 nan values to our dependent variables, it is better to drop the records instead of imputing.
```

```python
weather_data.dropna(axis=0,inplace = True)
#you have to pass entire data and not specific columns to drop the data. 
```

```python
weather_data.shape
```

```python
weather_data.isnull().sum()
```

```python
weather_data.head()
```

```python
#As Temp C and Temp is showing similar values we can drop 'Temp C'
```

```python
del weather_data['Temp C']
```

```python
weather_data.head()
```

```python
weather_data.dtypes
```

```python
#from dtypes we can see that Month feature is showing as object instead of integer.
```

```python
weather_data['Month'].unique()
```

```python
weather_data['Month'] = pd.to_numeric(weather_data['Month'],errors ='coerce')
```

```python
weather_data.head(2)
```

```python
weather_data['Month'].unique()
```

```python
weather_data['Month'].fillna(value = 5,inplace =True)
```

```python
weather_data.dtypes
```

```python
weather_data['Month'] = weather_data['Month'].astype('int64')
```

```python
weather_data.head()
```

```python
weather_data.dtypes
```

#Stage 2: Data Transformation

##Discrete Data

###Label Encoder

```python
weather_data_cleaned = weather_data.copy()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
weather_data_cleaned['weather_label_encoder']= le.fit_transform(weather_data_cleaned['Weather'])
```

```python
weather_data_cleaned.head(2)
```

```python
del weather_data_cleaned['Weather']
```

```python
weather_data_cleaned.head(2)
```

###One Hot Encoding

It can be done using 2 libraries:

  1. Pandas - get_dummies()
  
  2. sklearn - OneHotEncoder()

Using Pandas

```python
weather_data_cleaned_2 = weather_data.copy()
```

```python
weather_data_cleaned_2 = pd.get_dummies(data = weather_data_cleaned_2,columns = ['Weather'])
```

```python
weather_data_cleaned_2.head(2)
```

```python
weather_data_cleaned_2.isnull().sum()
```

Using Sklearn libary

```python
weather_data_cleaned_3 = weather_data.copy()
```

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np


ohe = OneHotEncoder()
ohe = pd.DataFrame(ohe.fit_transform(weather_data_cleaned_3[['Weather']]).toarray())
```

```python
ohe
```

```python
# merge with main df 
weather_data_cleaned_3final = weather_data_cleaned_3.iloc[:,0:4].join(ohe)
weather_data_cleaned_3final
```

```python
weather_data_cleaned_3final.isnull().sum()
```
