#Import library/data

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
Univ = pd.read_csv("Universities.csv")
```

```python
Univ
```

#Normalization function 

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_Univ_df = scaler.fit_transform(Univ.iloc[:,1:])
```

```python
scaled_Univ_df
```

```python
scaled_Univ_df = pd.DataFrame(scaled_Univ_df)
```

```python
# How to find optimum number of  cluster
#The K-means algorithm aims to choose centroids that minimise the inertia, or within-cluster sum-of-squares criterion:
```

#Cluster Validation Techniques

##Elbow Method

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i,random_state=0) #you can use any number in random_state
    kmeans.fit(scaled_Univ_df)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

##Silhouette Score

```python
from sklearn.metrics import silhouette_score as sil  #ranges between -1 to +1 and if the values is closer to +1 we say it as optimal number.
```

```python
import numpy as np
```

```python
wcss = []
for i in range(2, 15):   #silhouette_score requires more than 1 cluster labels. This causes the error. The error message is clear. Thus you cannot loop from 1
    kmeans = KMeans(i) 
    y = kmeans.fit_predict(scaled_Univ_df)
    wcss.append(sil(scaled_Univ_df,y))
    
plt.plot(range(2, 15), wcss)
plt.title('Silhouette Score Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

##Calinski harabaz score

```python
from sklearn.metrics import calinski_harabasz_score as chs
```

```python
wcss = []
for i in range(2, 15):   #requires more than 1 cluster labels. This causes the error. The error message is clear. Thus you cannot loop from 1
    kmeans = KMeans(i) 
    y = kmeans.fit_predict(scaled_Univ_df)
    wcss.append(chs(scaled_Univ_df,y))
    
plt.plot(range(2, 15), wcss)
plt.title('Calinski harabaz score Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
#maximum value to be taken as cluster
```

#Build Cluster algorithm

```python
from sklearn.cluster import KMeans
clusters_new = KMeans(4, random_state=42) #you can use any number in random_state
clusters_new.fit(scaled_Univ_df)
```

```python
#Assign clusters to the data set
Univ['clusterid_new'] = clusters_new.labels_
```

```python
#these are standardized values.
clusters_new.cluster_centers_
```
